```python
# %% [code]


# %% [code]
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np 
import librosa
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier

if __name__=='__main__':
    import os
    # for dirname, _, filenames in os.walk('~/Desktop/data'):
    #     for filename in filenames:
    #         print(os.path.join( filename))

    lst=[]
    for subdir, dirs, files in os.walk('~/Desktop/data'):
      for file in files:
           if (int(file[18:20])%2==0):
              try:
                #Load librosa array, obtain mfcss, store the file and the mcss information in a new array
                X, sample_rate = librosa.load(os.path.join(subdir,file), res_type='kaiser_fast')
                mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T,axis=0) 
                emotion = int(file[6:8])
                intensity=int(file[9:11])
                if emotion == 1:
                    continue
                if emotion == 6 or (emotion==4 and intensity==2) :
                    arr = mfccs, 1
                else:
                    arr=mfccs,0
                lst.append(arr)
              # If the file is not valid, skip it
              except ValueError:
                continue

    X, y = zip(*lst)

    X = np.asarray(X)
    y = np.asarray(y)


    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)


    randomforest = RandomForestClassifier(criterion="gini", max_depth=10, max_features="log2", 
                                     max_leaf_nodes = 100, min_samples_leaf = 3, min_samples_split = 20, 
                                     n_estimators= 2200, random_state= 5)

    randomforest.fit(X_train, y_train)
    from sklearn.metrics import classification_report
    predictions = randomforest.predict(X_test)   
    print(classification_report(y_test,predictions))
    
    


```
